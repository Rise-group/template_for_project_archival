```python
# Read image (check the path and the filename).
img1_BRG = cv2.imread('/template_for_project_archival/data/lena.bmp')

# Display basic information about the image.
print('Image dimensions:',img1_BRG.shape)
print('Image type:',img1_BRG.dtype)

#Convert BGR format from OpenCV to RGB format used my matplotlib.
img1_RGB =cv2.cvtColor(img1_BRG, cv2.COLOR_BGR2RGB)
plt.title('Lena')
plt.axis("off")
plt.imshow(img1_RGB)
plt.show()

#Use a 15x15 Gaussian kernel with standard deviation of 7.
img_blur_RGB =cv2.GaussianBlur(img1_RGB,(15,15),7) 
plt.title('Lena with some blur')
plt.axis("off")
plt.imshow(img_blur_RGB)
plt.show()
```
